```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

```python
dataset=pd.read_csv('Pre_processdata.csv')
```

```python
dataset
```

```python
datas=pd.get_dummies(dataset,drop_first=True)
```

```python
datas
```

```python
datas.columns
```

```python
ind=datas.drop('gender_Male',1)
dep=datas['gender_Male']
```

```python
ind
```

```python
dep
```

```python
def select_k(ind,dep,n):
    selection=SelectKBest(score_func=chi2,k=n)
    fit_select=selection.fit(ind,dep)
    select_feature=fit_select.transform(ind)
    return select_feature
```

```python
def split_scaler(ind,dep):
    X_train, X_test, y_train, y_test = train_test_split(ind, dep, test_size = 0.25, random_state = 0)
   
    return X_train, X_test, y_train, y_test
```

```python
def accuracy_pred(classifier,X_test):
    y_pred=classifier.predict(X_test)
    from sklearn.metrics import confusion_matrix
    cm=confusion_matrix(y_test,y_pred)
    from sklearn.metrics import classification_report
    clf=classification_report(y_test,y_pred)
    from sklearn.metrics import accuracy_score
    accuracy=accuracy_score(y_test,y_pred)
    return classifier,X_test,y_test,cm,clf,accuracy
def logis(X_train,y_train,X_test):
    from sklearn.linear_model import LogisticRegression
    classifier=LogisticRegression(random_state = 0)
    classifier.fit(X_train,y_train)
    classifier,X_test,y_test,cm,clf,accuracy=accuracy_pred(classifier,X_test)
    return classifier,X_test,y_test,cm,clf,accuracy
def random(X_train,y_train,X_test):
    from sklearn.ensemble import RandomForestClassifier
    classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
    classifier.fit(X_train,y_train)
    classifier,X_test,y_test,cm,clf,accuracy=accuracy_pred(classifier,X_test)
    return classifier,X_test,y_test,cm,clf,accuracy
def decision(X_train,y_train,X_test):
    from sklearn.tree import DecisionTreeClassifier
    classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
    classifier.fit(X_train,y_train)
    classifier,X_test,y_test,cm,clf,accuracy=accuracy_pred(classifier,X_test)
    return classifier,X_test,y_test,cm,clf,accuracy
def knn(X_train,y_train,X_test):
    from sklearn.neighbors import KNeighborsClassifier
    classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
    classifier.fit(X_train,y_train)
    classifier,X_test,y_test,cm,clf,accuracy=accuracy_pred(classifier,X_test)
    return classifier,X_test,y_test,cm,clf,accuracy
def navie(X_train,y_train,X_test):
    from sklearn.naive_bayes import GaussianNB
    classifier = GaussianNB()
    classifier.fit(X_train,y_train)
    classifier,X_test,y_test,cm,clf,accuracy=accuracy_pred(classifier,X_test)
    return classifier,X_test,y_test,cm,clf,accuracy
def tables(acclogis,accran,accdeci,accknn,accnavie):
    table=pd.DataFrame(index=['chi2'],columns=['acclogis','accran','accdeci','accknn','accnavie'])
    for number,i in enumerate(table.index):
        table['acclogis'][i]=acclogis[number]
        table['accran'][i]=accran[number]
        table['accdeci'][i]=accdeci[number]
        table['accknn'][i]=accknn[number]
        table['accnavie'][i]=accnavie[number]
    return table
```

```python
kbest=select_k(ind,dep,3)
```

```python
X_train, X_test, y_train, y_test=split_scaler(kbest,dep)
```

```python
acclogis=[]
accran=[]
accdeci=[]
accknn=[]
accnavie=[]
classifier,X_test,y_test,cm,clf,accuracy=logis(X_train,y_train,X_test)
acclogis.append(accuracy)
classifier,X_test,y_test,cm,clf,accuracy=random(X_train,y_train,X_test)
accran.append(accuracy)
classifier,X_test,y_test,cm,clf,accuracy=decision(X_train,y_train,X_test)
accdeci.append(accuracy)
classifier,X_test,y_test,cm,clf,accuracy=knn(X_train,y_train,X_test)
accknn.append(accuracy)
classifier,X_test,y_test,cm,clf,accuracy=navie(X_train,y_train,X_test)
accnavie.append(accuracy)
```

```python
result=tables(acclogis,accran,accdeci,accknn,accnavie)
```
